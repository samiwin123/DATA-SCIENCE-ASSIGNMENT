29. Solve the system of linear equations 2x + 3y = 5 and 4x + 6y = 10 using NumPY

```python
import numpy as np

# Coefficient matrix A
A = np.array([[2, 3],
              [4, 6]])

# Constants vector B
B = np.array([5, 10])

# Try solving the system
try:
    X = np.linalg.solve(A, B)
    print("Unique solution:", X)
except np.linalg.LinAlgError as e:
    print("No unique solution:", e)
```

30. Perform Singular Value Decomposition (SVD) on a 3x3 matrix and reconstruct the original
matrix using the SVD componen

```python
A = np.array([[3, 1, 1],
              [-1, 3, 1],
              [0, 2, 4]])

print("Original Matrix A:")
print(A)

# Step 2: Perform SVD
U, S, VT = np.linalg.svd(A)

print("\nU Matrix:")
print(U)

print("\nSingular values (as 1D array):")
print(S)

print("\nV Transpose (V^T):")
print(VT)

# Step 3: Convert S into a diagonal matrix (same shape as A)
S_matrix = np.zeros_like(A, dtype=float)
np.fill_diagonal(S_matrix, S)

# Step 4: Reconstruct A from SVD components
A_reconstructed = U @ S_matrix @ VT

print("\nReconstructed Matrix A:")
print(A_reconstructed)

# Step 5: Compare (optional)
print("\nIs close to original? →", np.allclose(A, A_reconstructed))
```
